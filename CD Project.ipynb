```python
accountscountry_data_table_create = (""" CREATE TABLE IF NOT EXISTS accountscountry(
Code VARCHAR PRIMARY KEY,
Long_Name VARCHAR,
Income_Group VARCHAR, 
Region VARCHAR, 
Lending_category VARCHAR
);""")
```
